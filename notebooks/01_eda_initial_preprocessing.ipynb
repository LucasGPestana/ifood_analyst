Todas os registros que possuem valores nulos em `Income` se referem a clientes que não aceitaram as ofertas das campanhas 2 e 3 (`AcceptedCmp2` e `AcceptedCmp3` iguais a 0). 

```python
# Média dos clientes que não aceitaram as ofertas das campanhas 2 e 3
df[(df["Income"].notnull()) & (df["AcceptedCmp2"] == 0) & (df["AcceptedCmp2"] == 0)]["Income"].median()
```

Usaremos a mediana de `Income` desses registros para imputar os valores ausentes de `Income`, para que o valor não seja sensível a outliers

```python
# Imputando a mediana de income para registros com essas características

median_income = df[(df["Income"].notnull()) & (df["AcceptedCmp2"] == 0) & (df["AcceptedCmp2"] == 0)]["Income"].median()
df.loc[df["Income"].isnull(), "Income"] = median_income
```

Não parece haver problemas de padronização de datas, nem datas errôneas em `Dt_Customer`, e também nenhuma inconsistência com `Year_Birth`

No entanto, o algoritmo de classificação pode ter problemas com relação ao tipo do atributo `Dt_Customer` ser datetime. Logo, é interessante gerar atributos derivados desse atributo.

Vou gerar um atributo numérico referente aos dias desde a inscrição do cliente (`DaysSinceEnrolled`) e um atributo categórico referente aos anos desde a inscrição do cliente (`YearsSinceEnrolled`), ambos com referência a data máxima de `Dt_Customer`.

```python
# Gerando o atributo DaysSinceEnrolled a partir de Dt_Customer
df["DaysSinceEnrolled"] = (df["Dt_Customer"].max() - df["Dt_Customer"]).dt.days
df["DaysSinceEnrolled"].head()
```

```python
# Gerando o atributo YearsSinceEnrolled a partir de Dt_Customer
df["YearsSinceEnrolled"] = df["DaysSinceEnrolled"] // 365
df["YearsSinceEnrolled"].head()
```

```python
# Verificando a cardinalidade de YearsSinceEnrolled
df["YearsSinceEnrolled"].nunique()
```

Também vou gerar um atributo derivado de `Year_Birth`, correspondente a idade do cliente em relação ao ano da data máxima de inscrição. Nesse caso, o algoritmo não vai ter tantos ganhos, mas pode ser mais interpretável para uma pessoa.

```python
# Gerando o atributo Age a partir de Year_Birth
df["Age"] = df["Dt_Customer"].max().year - df["Year_Birth"]
df["Age"].head()
```

```python
# Verificando se há registros duplicados
df.duplicated().sum()
```

```python
# Verificando a quantidade de registros da classe minoritária em Response
df[(df.duplicated()) & (df["Response"] == 1)].shape[0]
```

Registros duplicados não é necessariamente um problema para o algoritmo de classificação. Possivelmente, eles podem mascarar as métricas de avaliação do modelo, mas reduzir ainda mais a quantidade de registros da classe minoritária é pior ainda.

No entanto, registros duplicados com rótulos no atributo alvo inconsistentes será um grave problema, pois o algoritmo pode capturar os padrões de ambos os registros no treino e classificar erroneamente no teste. Esses registros duplicados, em especifico, serão removidos.

```python
# Verificando os registros duplicados que apresentam rótulos diferentes (inconsistência)
df.groupby(
    df.columns.drop("Response").tolist()
    ).filter(
        lambda x: x["Response"].nunique() >= 2
        )
```

```python
# Armazenando os indices registros duplicados com rótulos inconsistentes
inconsistences_indexes = df.groupby(
    df.columns.drop("Response").tolist()
    ).filter(
        lambda x: x["Response"].nunique() >= 2
        ).index
inconsistences_indexes
```

```python
# Removendo esses registros duplicados inconsistentes
df = df[~(df.index.isin(inconsistences_indexes))].reset_index(drop=True)
```

```python
# Verificando se ainda há registros duplicados inconsistentes
df.groupby(
    df.columns.drop("Response").tolist()
    ).filter(
        lambda x: x["Response"].nunique() >= 2
        )
```

```python
# Verificando a proporção do atributo alvo
df["Response"].value_counts(normalize=True)
```
